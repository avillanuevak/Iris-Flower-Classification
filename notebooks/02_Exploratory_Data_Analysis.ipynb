# Iris Flower Classification: Exploratory Data Analysis (EDA)

In this notebook, we will explore the Iris dataset. Our goal is to understand the relationships between the different features and how they relate to the target variable (the flower species)

### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

### Load the Dataset

```python
df = pd.read_csv('../data/iris.csv')
df.head()
```

### Correlation Heatmap

A heatmap is a way to visualize the correlation between our numerical features. Correlation tells us how strongly features are related to each other. A high positive correlation (close to 1) means that as one feature increases, the other tends to increase. A high negative correlation (close to -1) means that as one feature increases, the other tends to decrease.

```python
corr_matrix = df.drop('target', axis=1).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

**Observation from the Heatmap:**

We can see a strong positive correlation between `petal length` and `petal width` (0.96), and between `petal length` and `sepal length` (0.87). This confirms what we saw in the pair plot: these features tend to increase together. This high correlation is a key reason why models can effectively distinguish between the species.

### Pair Plot

```python
sns.pairplot(df, hue='target', palette='viridis')
plt.show()
```

### 3D Scatter Plot

```python
import plotly.express as px
fig = px.scatter_3d(df, x='sepal length (cm)', y='sepal width (cm)', z='petal width (cm)', color='target', title='3D Scatter Plot of Iris Dataset')
fig.show()
```
